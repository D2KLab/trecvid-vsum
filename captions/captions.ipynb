```python
ep2id = {'5531550228324592939.mp4': '175',
         '5534228999422914578.mp4': '176',
         '5539381671692122744.mp4': '177',
         '5542003749222140011.mp4': '178',
         '5544574287152993687.mp4': '179',
         '5544620672795594434.mp4': '180',
         '5547193787702629969.mp4': '181',
         '5549784941472309008.mp4': '182',
         '5552368364300855101.mp4': '183',
         '5555325449284154780.mp4': '184',
         '5555360238519252381.mp4': '185'}

id2ep = {v:k for k,v in ep2id.items()}
```

```python
data = []
with open('captions.txt') as f:
    for line in f:
        ep = line[26:29]
        shot = line.split(':s')[1].split(' ')[0]
        caption = line[55:].strip()
        assert(int(ep) in range(175, 186))
        assert(int(shot))
        data.append((id2ep[ep], int(shot), caption))
        
df = pd.DataFrame(data, columns=['episode', 'shot', 'caption'])
```
